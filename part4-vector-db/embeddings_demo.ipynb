# Part 4.1 — Vector Embeddings Demo
In this notebook, we demonstrate how to generate vector embeddings for products and calculate their semantic similarity.

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer, util
import numpy as np

# 1. Load a pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# 2. Select 5 products from our dataset
products = [
    "Wireless Bluetooth Headphones",
    "Noise Cancelling Earbuds",
    "Ergonomic Office Chair",
    "Standing Desk with Motor",
    "Mechanical Gaming Keyboard"
]

# 3. Generate Vector Embeddings
embeddings = model.encode(products)
print(f"Generated {len(embeddings)} embeddings. Each vector has {len(embeddings[0])} dimensions.")
```

```python
# 4. Calculate Cosine Similarity
# Pair 1: Similar products (Headphones and Earbuds)
sim_pair = util.cos_sim(embeddings[0], embeddings[1])

# Pair 2: Dissimilar products (Headphones and Office Chair)
diff_pair = util.cos_sim(embeddings[0], embeddings[2])

print(f"Similarity (Headphones vs Earbuds): {sim_pair.item():.4f}")
print(f"Similarity (Headphones vs Office Chair): {diff_pair.item():.4f}")
```

### Recommended Products Feature Explanation
These embeddings can be used to build a **'Recommended Products'** feature by storing all product vectors in a Vector Database. 

When a user views a product (e.g., 'Headphones'), the system calculates the **Cosine Similarity** between that product's vector and all other products in the store. The items with the highest similarity scores (closest in vector space) are then displayed as 'Similar Items' or 'You might also like.' Unlike traditional SQL, this works even if the product names don't share any exact keywords, because the model understands the **meaning** of the words.
